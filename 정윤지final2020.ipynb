## [문제1] (24점)

```python
# class Mail 을 구현하시오.[12점] Implement class Mail

class Mail:
    def __init__(self, sAdd, rAdd, express = False):
        self.sAdd = sAdd
        self.rAdd = rAdd
        self.express = express
        
    def getCost(self):
        if self.express == True:
            return 1000
        else:
            return 500
        
    def __str__(self):
        if self.express == False:
            rep = "Mail from {0} to {1} {2} won.".format(self.sAdd, self.rAdd, self.getCost())
            return rep
        else:
            rep = "Express Mail from {0} to {1} {2} won.".format(self.sAdd, self.rAdd, self.getCost())
            return rep
```

```python
# class Mail을 상속받아 class Parcel 을 구현하시오.[12점] Implement class Parcel which inherits class Mail

class Parcel(Mail):
    def __init__(self, sAdd, rAdd, weight, express = False):
        super().__init__(sAdd, rAdd, express)
        self.weight = weight
        
    def getCost(self):
        if self.express == False:
            if self.weight < 1.0:
                return 3000
            elif 1.0 <= self.weight < 4.0:
                return 4000
            else:
                return 5000
        else:
            if self.weight < 1.0:
                return 4000
            elif 1.0 <= self.weight < 4.0:
                return 5000
            else:
                return 6000
            
        
    def __str__(self):
        if self.express == False:
            rep = "Parcel {0} kg from {1} to {2} {3} won.".format(self.weight, self.sAdd, self.rAdd, self.getCost())
            return rep
        else:
            rep = "Express Parcel {0} kg from {1} to {2} {3} won.".format(self.weight, self.sAdd, self.rAdd, self.getCost())
            return rep
```
