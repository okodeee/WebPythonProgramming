# 마피아 게임

프로젝트 목표 및 내용:
 마피아, 경찰, 의사, 시민으로 구성된 기본 마피아 게임의 진행을 돕는다. 

주제 선정 이유 또는 이 프로젝트의 필요성:
 마피아 게임에서 발생하는 불편을 보완하기 위해 이 주제를 선정했다.
 일반적으로 마피아 게임은 진행자가 한 명 필요하다. 진행자는 플레이어들에게 직업을 부여하고 결과를 공지하는 절대적인 역할이다. 따라서 진행자가 실수를 하거나 진행이 서툰 경우 게임의 재미가 줄어든다. 또한 진행자는 실질적으로 게임에 참여하지 않기 때문에 큰 흥미를 느끼기 어렵다. 게임의 재미를 반감시키는 것에는 물리적인 요인도 있다. 마피아 게임은 플레이어들끼리의 토론 시간, 개별적으로 갖는 진행자와의 소통 시간이 있다. 진행자와 접촉하면서 발생하는 바람, 옷 소리로 그 대상을 알아차릴 수 있다. 이는 추리를 하는 플레이어들을 허탈하게 만든다.  
 이 프로그램으로 마피아 게임을 진행한다면 재미가 극대화될 것이다.

구현 내용 설명:
 인원 수를 입력받은 후 랜덤 함수로 각 플레이어들의 이름에 역할을 부여한다.
 (5명 진행 시: 마피아 1명, 의사 1명, 경찰 1명, 시민 2명 ...)
 낮 동안 타이머를 작동시켜 플레이어들이 토론을 진행하게 한다. 
 플레이어들이 투표로 탈락시킨 사람의 정체를 알려준다.
 밤이 되면 각자 능력을 사용하게 한다.
 (경찰은 이름을 입력하면 그 사람의 정체를 알려주고, 의사와 마피아의 지목에 따라 탈락할 사람을 계산한다.)
 다음 날이 되면 밤 동안 탈락한 사람을 알려주고 정체를 공개한다.
 반복문을 사용하여 마피아의 수가 시민의 수보다 많거나 같으면 마피아의 승으로,
 마피아가 모두 탈락하면 시민의 승으로 게임을 종료시킨다.
